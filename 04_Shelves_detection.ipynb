```python
def move_point_inside(data, min_x, max_x, min_y, max_y):
    thrs = 0.5
    x = data["x"].tolist()
    y = data["y"].tolist()
    for i in range(len(x)):
        if x[i] <= min_x:
            x[i] = min_x + thrs
        if x[i] >= max_x:
            x[i] = max_x - thrs
        
        if y[i] <= min_y:
            y[i] = min_y + thrs
        if y[i] >= max_y:
            y[i] = max_y - thrs
   
    new_data = data
    new_data["x"] = x
    new_data["y"] = y
    
    return new_data

def move_point(polygon, data):
    thrs = 0.5
    bounds = polygon.bounds

    x = []
    y = []

    if len(data) == 1:
        point_x = data["x"].iloc[0]
        point_y = data["y"].iloc[0]

        # min_x
        dist_x1 = abs(point_x - bounds[0])

        # max_x
        dist_x2 = abs(point_x - bounds[2])

        # min_y
        dist_y1 = abs(point_y - bounds[1])

        # max_y
        dist_y2 = abs(point_y - bounds[3])

        value = [dist_x1, dist_x2, dist_y1, dist_y2]
        min_idx = value.index(min(value))

        case = {
            0: bounds[0] - thrs,
            1: bounds[2] + thrs,
            2: bounds[1] - thrs,
            3: bounds[3] + thrs,
        }

        if min_idx == 0 or min_idx == 1:
            x.append(case[min_idx])
            y.append(point_y)
        else:
            x.append(point_x)
            y.append(case[min_idx])
    else:
        std_y = data["y"].std()
        std_x = data["x"].std()

        if std_y >= std_x:
            #si sposta tanto in asse y, quindi vanno corrette le x
            mean_x = data["x"].mean()

            # min_x
            dist_1 = abs(bounds[0] - mean_x)

            # max_x
            dist_2 = abs(bounds[2] - mean_x)

            if dist_1 <= dist_2:
                data["x"] = bounds[0] - thrs
            else:
                data["x"] = bounds[2] + thrs

        else:
            mean_y = data["y"].mean()

            # min_y
            dist_1 = abs(bounds[1] - mean_y)

            # max_y
            dist_2 = abs(bounds[3] - mean_y)

            if dist_1 <= dist_2:
                data["y"] = bounds[1] - thrs
            else:
                data["y"] = bounds[3] + thrs

        x += data["x"].tolist()
        y += data["y"].tolist()
    return x, y

"""
Per i scaffali 34 35 36 37 38 39 40 41 che sono nei mordi del supermercato
da applicare a shelve_32 shelve_33 shelve_34
"""
def move_point_custom(polygon, data):
    thrs = 0.5
    bounds = polygon.bounds

    x = []
    y = []

    if len(data) == 1:
        point_x = data["x"].iloc[0]
        point_y = data["y"].iloc[0]

        # max_x
        dist_x = abs(point_x - bounds[2])

        # max_y
        dist_y = abs(point_y - bounds[3])

        value = [dist_x, dist_y]
        min_idx = value.index(min(value))

        case = {
            0: bounds[2] + thrs,
            1: bounds[3] + thrs,
        }

        if min_idx == 0:
            x.append(case[min_idx])
            y.append(point_y)
        else:
            x.append(point_x)
            y.append(case[min_idx])
    else:
        std_y = data["y"].std()
        std_x = data["x"].std()

        if std_y >= std_x:
            #si sposta tanto in asse y, quindi vanno corrette le x
            # max_x
            data["x"] = bounds[2] + thrs

        else:
            # max_y
            data["y"] = bounds[3] + thrs

        x += data["x"].tolist()
        y += data["y"].tolist()
    return x, y

"""
Per i scaffali 42 43 44 che sono nei mordi del supermercato
da applicare a shelve_35 shelve_44
"""
def move_point_custom_2(polygon, data):
    thrs = 0.5
    bounds = polygon.bounds

    x = []
    y = []

    if len(data) == 1:
        point_x = data["x"].iloc[0]
        point_y = data["y"].iloc[0]

        # min_x
        dist_x = abs(point_x - bounds[0])

        # min_y
        dist_y = abs(point_y - bounds[1])

        value = [dist_x, dist_y]
        min_idx = value.index(min(value))

        case = {
            0: bounds[2] + thrs,
            1: bounds[3] + thrs,
        }

        if min_idx == 0:
            x.append(case[min_idx])
            y.append(point_y)
        else:
            x.append(point_x)
            y.append(case[min_idx])
    else:
        std_y = data["y"].std()
        std_x = data["x"].std()

        if std_y >= std_x:
            #si sposta tanto in asse y, quindi vanno corrette le x
            # min_x
            data["x"] = bounds[0]- thrs

        else:
            # min_y
            data["y"] = bounds[1]-thrs

        x += data["x"].tolist()
        y += data["y"].tolist()
    return x, y
```
